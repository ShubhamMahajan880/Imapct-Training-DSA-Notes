# **Circular Link List** :-
Circular Link List  will be similar to single link list but in the Single Link List the last node address will be null but in the circular link list last node will be connected to the first node,that's why last node address is address of the first node.

* Single Link List contains only head but circular link list contains Head & Tail

* Head always pointing to first node

* Tail always pointing to last node.1.jpg

Qun :-**WAP to create 4 nodes in the circular link list and display the data present in the link list and reverse the data present in the nodes.**

2.jpg

```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None

class cll:
  def __init__(self):
    self.head=None
    self.tail=None

  def display(self):
    if(self.head is None):
      print("Empty")
    else:
      temp=self.head
      while(True):
        print(temp.data,end="----->")
        temp=temp.next
        if(temp==self.head):
          break
      print(temp.data)


n1=Node(10)
l1=cll()
l1.head=n1
l1.tail=n1
l1.tail.next=l1.head
n2=Node(20)
l1.tail.next=n2
l1.tail=n2
l1.tail.next=l1.head
n3=Node(30)
l1.tail.next=n3
l1.tail=n3
l1.tail.next=l1.head
n4=Node(40)
l1.tail.next=n4
l1.tail=n4
l1.tail.next=l1.head
l1.display()

```
