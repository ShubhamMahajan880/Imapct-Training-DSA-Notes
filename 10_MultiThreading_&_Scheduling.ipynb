# **Multi Threading** : -

 * Thread is a functionality or logic which executes simultaneously along with the other part of the program.

* Thread is a light weight process.

* A program which is under execution is called as a process.

* We can define functionality or logic as a thread by overiding run() method of thread class.

* Thread class is predefined class which is defined in threading module.

* Threads can be executed through start() method of thread class.




```python
import threading
class x(threading.Thread):
  def run(self):
    for p in range(1,101):
      print(p)

class y(threading.Thread):
  def run(self):
    for q in range(101,201):
      print(q)


x1=x()
x1.start()
y1=y()
y1.start()

```

# **Thread Name** : -

* Python interpreter assign one unique name to each and every thread.

* We can get names of thread by calling getname method of Thread class.

* We caan also assign our own names to threads by calling SetName() method of thread class.

```python
import threading
class x(threading.Thread):
  def run(self):
    print(self.getName())

class y(threading.Thread):
  def run(self):
    print(self.getName())


x1=x()
x1.start()
y1=y()
y1.start()


# using getName() method opf thread classs we got the name of thread by defualt
# Output is very like Thread-1,Thread-2 or Thread-17,Threaad-16 .. it totally depends on Thread Scheduler of System by Default
```

```python
import threading
class x(threading.Thread):
  def run(self):
    print(self.getName())

class y(threading.Thread):
  def run(self):
    print(self.getName())


x1=x()
x1.setName("XXX-Thread")
x1.start()
y1=y()
y1.setName("YYY-Thread")
y1.start()


# using setName() method opf thread classs we define the name of thread according to we want

# XXX-Thread, YYY-Thread
```

# **Suspendiung the Execution of Thread** : -




* We can suspend Execution of thread temporarly by calling Sleep function or Join Method.

* Sleep function is a predefined function which is defined in time module.

* Sleep function suspense the execution of current thread untill specified time is over.

```python
import time
for p in range(1,21):
  print(p)
  time.sleep(2)

  # Using the sleep method we can observe the printing of output for specific sleep time period
```

# **Join Method** : -

* Join method is a pre defined method, which is defined in thread class

* Join Method stop the execution of current thread until speficifed thread execution is over.



```python
import time
import threading
class y(threading.Thread):
  def run(self):
    for q in range(1,6):
      print("Seetha thread")
      time.sleep(2)

y1=y()
y1.start()
y1.join() # y1 will join after the execution of Seeta Thread
#y1.join(5) # After 5 seconds y1 that is Rama Thread actually will join immediately

for p in range(1,11):
  print("Rama Thread")
```
