# **Nested while Loop** : -

A while Loop within the another while is called as Nested While loop

Qun :- **WAP to print Prime Numbers between 1 to 100 using Nested while**


```python
l=[]
i=2
while(i<=100):
  c=0
  q=1
  while(q<=i):
    if(i%q==0):
      c=c+1
    q=q+1
  if(c==2):
    l.append(i)
  i=i+1
print(l)
print(len(l))
```
