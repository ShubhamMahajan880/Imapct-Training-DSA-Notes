# **Double Linked List**

Double Linked List is also collection of nodes.


* and every node will be connected with the next node.

* In Double Linked List every niode will contain 3 fields :-

1 - Previous(prev preveous node address)

2 - data()

3 - next()

* For the first node there is no any previous node therefor its null


* For the last node next field address will be null

* the basic operations we can perform on the Double Linked List are ðŸ‡±

i - Creating the Nodes

ii - Inserting the data into the nodes

iii- Display the data present in the node

iv - Reversing the data present in the node thorugh the tail

v - adding the nodes

vi - Deleting the nodes 5.jpg
