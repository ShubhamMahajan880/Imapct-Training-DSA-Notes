# **Linear Search** :-

In the Linear Search we are going to compare key element with the each and every element present in the list

* If the element found then it will return the index possition of the corresponding element

* If the element is not available then it will return -1, that is -1 indicated as element not found

* In the linear search, we are going to compare each and every element with the key so this algorithm will take more time and time complexity also more



```python
def linear_search(l,key):#[12,23,34,67,45,52,78],45
  for p in range(len(l)):
    if(l[p]==key):
      return p
  return -1

l=list(map(int,input().split()))#[12,23,34,67,45,52,78]
key=int(input())
s=linear_search(l,key)
if(s==-1):
  print("Element Not FOund")
else:
  print("Element found at:",s,"index Possition")
```

# **Binary Search**  :-

* Step - 1 : -  In the binary search , first we have to sort the element

* Step - 2 : - Calculate Lower and High Values that is Low = 0 and High = len(n)-1

* Step - 3 : - Calcualate Middle Next Possition that is (Low + High)/2

* Step - 4 : - Based on the mid possition findout middle value in the list then compare, middle value with the key value,

if Middle value < key Value then Increment the Low

if Middle Value > Key Value then decrement the High with 1

If two conditions are not satiesfied that will be indicated as Middle Value & Key Value both are EQUAL

then return the index possition.

* If the element is not available then this algorithm  will return -1

```python
def selection_sort(l):
  for p in range(len(l)):
    min_index=p
    for q in range(p+1,len(l)):
      if(l[min_index]>l[q]):
        min_index=q
    l[min_index],l[p]=l[p],l[min_index]
  return l

l=list(map(int,input().split()))
print("Before Sorting : ",l)
s=selection_sort(l)
print("After Sort :",s)

```

# **Merge Shot** :-

* Merge shot works on the principle of divide & Conquer technique

* First we have to divide the element into two equal parts that is len(n)//2 ,
this division is possible untill the length = 1.

* for dividing the elements into equal parts we are using recursion technique

* During Comparision we are going to compare Left Side value with the Right Side value,
if Left Side Value less than(<) Right Side Value, then left side value will come first

If condition false, right side value will come first

```python
def merge(left,right):
  merged=[]
  i=0
  j=0
  while((i<len(left))and (j<len(right))):
    if(left[i]<right[j]):
      merged.append(left[i])
      i=i+1
    else:
      merged.append(right[j])
      j=j+1
  merged=merged+left[i:]+right[j:]
  return merged

def merge_sort(l):
  if(len(l)==1):
    return l
  else:
    mid=len(l)//2
    left=merge_sort(l[:mid])
    right=merge_sort(l[mid:])
  return merge(left,right)

l=[40,60,30,50,10,20]
print("Before sorting : ",l)
s=merge_sort(l)
print("After Sorting :",s)
```

# **Insertion Sort**  : -

```python
def insertion_sort(l):
  for p in range(1,len(l)):
    key=l[p]
    q=p-1
    while((q>=0) and (key<l[q])):
      l[q+1]=l[q]
      q=q-1
    l[q+1]=key
    print(l)

  return l

l=[5,4,1,2,3]
print("Before Sort : ",l)
r=insertion_sort(l)
print("After Sort: ",r)
```
