## **Basic Terminologies of a tree** : -

# 1 - **Node** : -

It is an element of a tree

a tree can have number of Nodes

# 2 - **Root Node** :-

* Starting element of tree is called as root nodes(where branch occurs)

* Tree wil have one root node

# 3 - **Edge** :-

* A link between One Noide to Another Node **OR** Connection between the Nodes

* If we have n nodes & Number of edges will be n-1

# 4 - **Parent** :-  

* A node which has branches is called as Parents

# 5 - **Child** :-

* Branches of a parent **OR** Node with edge from bottom to top

# 6 - **Siblings** :-

* The childrens of same parents

# 7 - **Leaf Node** :-

* A node wihtout any child is called as Leaf Node

# 8 - **Internal Nodes** :-

* All the nodes other than the leaf Node **OR** Node with child is called as "Internal Node"

# 9 - **Degree of a Tree** :-

* Number of child node will represent the degree of a Node

# 11 - **Height of a tree** :-

```python
def depth(root):
  if root is None:
    return 0
  ldepth=depth(root.left)
  rdepth=depth(root.right)
  print("Depth left:",ldepth)
  print("Depth right:",rdepth)
  if(ldepth>rdepth):
    return ldepth+1
  else:
    return rdepth+1


b=Node(27)
b.insert(14)
b.insert(35)
b.insert(10)
b.insert(19)
b.insert(31)
b.insert(42)
b.insert(127)
h=depth(b)
print("The Height if BST : ",h)


```
